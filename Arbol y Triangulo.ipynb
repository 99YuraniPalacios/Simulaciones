## Ejercicio 1: TRIÁNGULO DE SIRPINSKI
El triángulo de Sirpinski se forma al juntar los puntos medios de los lados de un triangulo equilátero y repetir este proceso con los triángulos resultantes. El ejercicio consiste en repro ducir la estructura de la Fig. 1. 

Los pasos para generar dicha estructura son:

* Definir las coordenadas de los tres vértices del triángulo equilátero.

* Definir un punto inicial desde el cual se empezará a trazar los demás puntos (le sugiero tomar uno de los puntos medios de cualquiera de los lados del triángulo equilátero).

* Generar un número aleatorio entre 1 y 3, tal que si el número es 1 el siguiente punto esté en el medio de la línea que une el primer vértice y su punto inicial, si el número es 2 el punto será el punto medio de la linea que une el punto inicial y el segundo vértice y finalmente si el número es 3 el punto será el punto medio que une la línea que une el punto y el tercer vértice. Ahora defina su nuevo punto de partida como el punto resultante del procedimiento anterior, al repetir un número adecuado de veces (el cual debe determinar), podrá obtener la Fig. 1.

```python
from  random  import randint
import matplotlib.pyplot as plt

#COORDENADAS DE LOS TRES VERTICES
Ax, Ay = 0.0, 0.0
Bx, By = 6.0, 0.0
Cx, Cy = 3.0, 5.2

# PUNTO INICIAL
Dx, Dy = (Ax+Bx)/2., (Ay+By)/2. #Punto medio entre A y B.

#NUMEROS ALEATORIOS:
ptox = []
ptoy = []

for i in range(50000):
    
    num = randint(1,3) # Tambien podemos desde random importar:
                       #             uniform, random. 
                       # num =int(uniform(1,4))
                       # num =int(1+random()*3)
    #PUNTO MEDIO ENTRE PUNTO INICIAL Y VERTICE A
    if num==1:
        ptox += [(Ax+Dx)/2.,] 
        ptoy += [(Ay+Dy)/2.,]
        Dx, Dy = ptox[i], ptoy[i]
        
    #PUNTO MEDIO ENTRE PUNTO INICIAL Y VERTICE B
    if num==2:
        ptox += [(Bx+Dx)/2.,] 
        ptoy += [(By+Dy)/2.,]
        Dx, Dy = ptox[i], ptoy[i]
    
    #PUNTO MEDIO ENTRE PUNTO INICIAL Y VERTICE C
    if num==3:
        ptox += [(Cx+Dx)/2.,] 
        ptoy += [(Cy+Dy)/2.,]
        Dx, Dy = ptox[i], ptoy[i] 
```

```python
%matplotlib inline
import matplotlib.pyplot as  plt
plt.plot(ptox, ptoy, ".")
```

El porcentaje dado es la probabilidad que se le asigna a cada punto, tal que parte importante de la solución es generar esta probabilidad. El resultado obtenido es dado en la Fig. 2

```python
from random import random

#PUNTO INICIAL
Ax, Ay = 0.0, 0.0

x=[]
y=[]

N = 5000
for i in range(N):
    n = random() #Creacion de numeros aleatorios entre 0 y 1.
    
    # Primera funcion que corresponde al 10% (0.1-0.0= 0.1)
    if n <= 0.1:
        x += [0.05*Ax]
        y += [0.6*Ay]
        Ax, Ay = x[i],y[i]
       
    # Segunda funcion que corresponde al 10% (0.2-0.1= 0.1)
    if 0.1 < n <= 0.2:
        x += [0.05*Ax]
        y += [-0.5*Ay+1.0]
        Ax, Ay = x[i],y[i]
    
    # Tercera funcion que corresponde al 20% (0.4-0.2= 0.2)
    if 0.2 < n <=0.4:
        x +=[0.46*Ax-0.15*Ay]
        y += [0.39*Ax+0.38*Ay+0.6]
        Ax, Ay = x[i],y[i]
        
    # Cuarta funcion que corresponde al 20% (0.6-0.4= 0.2)
    if 0.4 < n <=0.6:
        x += [0.47*Ax - 0.15*Ay]
        y += [0.17*Ax + 0.42*Ay +1.1]
        Ax, Ay = x[i],y[i]
    
    # Quinta funcion que corresponde al 20% (0.8-0.6= 0.2)
    if 0.6 < n <=0.8:
        x += [0.43*Ax + 0.28*Ay]
        y += [-0.25*Ax+0.45*Ay+1.0]
        Ax, Ay = x[i],y[i]
    
    # Sexta funcion que corresponde al 20% (1.0-0.8= 0.2)
    if n > 0.8:
        x += [0.42*Ax+0.26*Ay]
        y += [-0.35*Ax+0.31*Ay+0.7]
        Ax, Ay = x[i],y[i] 
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(x,y, '.', color = 'g')
```

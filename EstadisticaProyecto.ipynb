# PROBLEMA 2

Una central telefónica recibe en promedio λ llamadas por hora. Si el número de llamadas se distribuye según una Poisson y la central tiene una capacidad para atender a lo sumo 12 llamadas por minuto, entonces:

## PUNTO 1:
Construya un algoritmo que genera la distribución de probabilidad del caso de la central telefónica.


## SOLUCIÓN.

#### DEFINICION DE LAS VARIABLES

+ Lambda: Promedio de llamadas en un intervalo de tiempo (En minutos).
+ x: Variable aleatoria discreta que indica el numero de eventos que ocurre en un tiempo determinado.

En el caso de las llamadas lambda es igual al promedio de llamadas por horas y x es una variable aleatoria discreta que indica el numero de llamadas que ocurre en un tiempo determinado.

#### FUNCIÓN POISSON
Nuestra funcion Poisson calcula y devuelve la probabilidad de que ocurra un determinado número de eventos durante un tiempo instantantáneo, que a su vez puede cumplir la misma función que el comando `dpois(x, lambda, log=F)` en el lenguaje utilizado R.

#### FUNCIÓN APOISSON
Nuestra función APoisson calcula y devuelve la distribución de poisson acumulada, que en R se conoce con el comando `ppois(x, lambda, lower.tail = T, log.p = F)`.

#### FUNCIÓN DPOISSON
Nuestra función DPoisson calcula y devuelve un vector de probabilidades de x = (1, 2, 3, ... n), dada un n que indica la longitud del vector y lambda.

```r
Poisson <-function(x, lambda){
    return ((lambda^x)*exp(1)^-lambda/factorial(x))
    }

APoisson <- function(x, lambda){
    suma = 0
    for (i in 0:x){
        suma = suma + Poisson(i, lambda)
    }
    return(suma)
    }

DPoisson <- function(n, lambda){
    vector = numeric(n)
    for (i in 0:n){
        vector[i] = dpois(i, lambda)
    }
    return(vector)
    }
```

## PUNTO 2
Aplique el algoritmo anterior a cada uno de los siguientes valores de λ y gráfique cada caso. 
+ λ = 3
+ λ = 10
+ λ = 20
+ λ = 100
+ λ = 1000 

## SOLUCIÓN

```r
lambdas <- c(3, 10, 20, 100, 1000)
for (i in 1:5){
    x = seq(1, lambdas[i]*2)
    y = DPoisson(lambdas[i]*2, lambdas[i])
    plot(x,y, type="p", xlab=toString(c("Número de llamadas; λ =", lambdas[i])), ylab='P(x)')
}
```

## PUNTO 4
Si el número promedio de llamadas en una hora es de 480, ¿cuál es la probabilidad de que en un minuto determinado no sea posible dar línea a todos los clientes?

## SOLUCIÓN
El problema nos dice que:
 λ: 480 llamadas por hora que equivale a 8 llamadas por minuto.
 λ = 8.
 
 Se dice que en 1 minuto se hacen a lo sumo 12 llamadas, por lo tanto la probabilidad de que no sea posible dar línea a todos los clientes es cuando se sobrepasa este límite siendo:
 

```r
# P(X>12)= 1-P(X≤12) =
R = 1-APoisson(12, 8)
print("La probabiidad de que en un minuto determinado no sea posible dar línea a todos los clientes es de"); R
```
